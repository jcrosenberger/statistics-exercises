### 1. A bank found that the average number of cars waiting during the noon hour at a drive-up window follows a Poisson distribution with a mean of 2 cars. Make a chart of this distribution and answer these questions concerning the probability of cars waiting at the drive-up window.

- What is the probability that no cars drive up in the noon hour?
- What is the probability that 3 or more cars come through the drive through?
- How likely is it that the drive through gets at least 1 car?

```python
# Determine the statistical likelihood for the outcome of 0 cars waiting at noon considering an average of
# 2 cars 
zero_cars = stats.poisson(2).pmf(0)
print(f'On any given day, there is a {round(zero_cars*100,3)}% chance that no cars will be waiting at the drive-up window.')
```

```python
three_cars = stats.poisson(2).pmf(3)
print(f'On any given day, there is a {round(three_cars*100,3)}% chance that 3 cars will be waiting at the drive-up window.')
```

```python
at_least_1 = stats.poisson(2).cdf(1)
print(f'On any given day, there is a {round(at_least_1*100,3)}% chance that 0 or 1 cars will be waiting at the drive-up window.')
```

```python
# generate a random sample using our given mean. 10,000 observations, mean of 2 cars
x = poisson.rvs(mu=2, size= 10000)
```

```python
# create plot based on the random sample generated above
plt.hist(x, density=True, edgecolor = 'black')
```

### 2. Grades of State University graduates are normally distributed with a mean of 3.0 and a standard deviation of .3. Calculate the following:

- What grade point average is required to be in the top 5% of the graduating class?
- What GPA constitutes the bottom 15% of the class?
- An eccentric alumnus left scholarship money for students in the third decile from the bottom of their class. Determine the range of the third decile. Would a student with a 2.8 grade point average qualify for this scholarship?
- If I have a GPA of 3.5, what percentile am I in?

```python
mu = 3.0
sd = .3

grades=stats.norm(mu, sd)
```

```python
top_5gpa = grades.ppf(.95)
print(f'The top 5% of GPA is {round(top_5gpa,2)} or higher')
```

```python
bottom_15gpa = grades.ppf(.15)
print(f'The bottom 15% of GPA is {round(bottom_15gpa,2)} or lower')
```

```python
lower_30gpa = round(grades.ppf(.30),2)
lower_39gpa = round(grades.ppf(.39),2)
print(f'To be in the bottom decile of GPA, you would need to be above {lower_30gpa} but below {lower_39gpa}.')
print('A student with a gpa of 2.8 will not qualify for that scholarship.')
```

```python
high_gpa = round(grades.cdf(3.5)*100,2)
print(f'A student with a GPA of 3.5 is in the top {high_gpa} percentile')
```
